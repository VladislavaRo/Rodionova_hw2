2. (2 –±–∞–ª–ª–∞) –î–∞–Ω—ã –¥–≤–∞ —Å–ø–∏—Å–∫–∞. "–°–ª–∏—Ç—å" –∏—Ö —á–µ—Ä–µ–∑ –æ–¥–∏–Ω (—á–µ—Ä–µ–¥—É—è —ç–ª–µ–º–µ–Ω—Ç—ã —Ç–æ –æ–¥–Ω–æ–≥–æ, —Ç–æ –¥—Ä—É–≥–æ–≥–æ —Å–ø–∏—Å–∫–∞).

`f = [1,2,3]`

`s = ['1','2','3']`

–û–∂–∏–¥–∞–µ–º—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç:

`[1, '1', 2, '2', 3, '3']`


* _–°–ø–∏—Å–æ–∫ –º–æ–∂–µ—Ç –±—ã—Ç—å –ª—é–±–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞, –∞ –≤–∞—à –∫–æ–¥ –≤—Å–µ —Ä–∞–≤–Ω–æ –¥–æ–ª–∂–µ–Ω —Ä–∞–±–æ—Ç–∞—Ç—å._
* _–º–æ–∂–Ω–æ –∑–∞—Ä–∞–±–æ—Ç–∞—Ç—å + 2 –±–∞–ª–ª–∞ –∫ –∑–∞–¥–∞–Ω–∏—é, –µ—Å–ª–∏ —Å–¥–µ–ª–∞—Ç—å –µ–≥–æ –≤ –æ–¥–Ω—É —Å—Ç—Ä–æ–∫—É_

````python
#OK

x = []
f = [1,2,3,4,5,6,7,8,9]
s = ['1','2','3','4','5','6']

for i in range(max(len(f), len(s))):
    if i < len(f):
        x.append(f[i])
    if i < len(s):
        x.append(s[i])
print(x)
````

    - –ü—Ä–µ–≤—Ä–∞—Ç–∏—Ç–µ –µ–µ –≤ –ª–∏—Å—Ç —Å–∏–º–≤–æ–ª–æ–≤

````python
#OK

x = '–π—Ç—Ü—ã—É —Ñ–æ—á—Ç–º–ª–∑–∏—Ä—á–π–Ω–±–æ—é –∞—Å–∞–ø—ã—Ä–π–∞—Ñ–≤—è–ª–≥—è—à–µ—â—à–æ—å—Ä—Å–∫—è—É!–π'

y = list(x)
z = []

for i in range(len(y)):
  if i % 2 != 0:
    z.append(y[i])

print(''.join(map(str, z)), ', –∏ —ç—Ç–æ ', type(''.join(map(str, z))), '! –£—Ä–∞!')
````

38. (2 –±–∞–ª–ª–∞) –†–∞–∑–¥–µ–ª–∏—Ç–µ –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏—è –Ω–∞ —Å–ª–æ–≤–∞, —É–¥–∞–ª–∏—Ç–µ –ø—É–Ω–∫—Ç—É–∞—Ü–∏—é –¥–æ –∏ –ø–æ—Å–ª–µ —Å–ª–æ–≤, –∏—Å–ø–æ–ª—å–∑—É—è `.strip()`. –ò –∫–∞–≤—ã—á–∫–∏ —Ç–æ–∂–µ —É–¥–∞–ª–∏—Ç–µ.

````python
#OK

y = x.split('\n')

z = [i.split() for i in y]

from string import punctuation

a = [[n.strip(punctuation).strip('¬´¬ª') for n in i] for i in z]
print(a)
````

39. (2 –±–∞–ª–ª–∞) –°–æ–∑–¥–∞–π—Ç–µ —Å–ª–æ–≤–∞—Ä—å, –≤ –∫–æ—Ç–æ—Ä–æ–º –∫–ª—é—á -- —Å–ª–æ–≤–æ –∏–∑ —Ç–µ–∫—Å—Ç–∞, –∞ –∑–Ω–∞—á–µ–Ω–∏–µ -- –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ä–∞–∑, —Å–∫–æ–ª—å–∫–æ –æ–Ω–æ –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è –≤–æ –≤—Å–µ–º —Ç–µ–∫—Å—Ç–µ.

–Ω–∞–ø—Ä–∏–º–µ—Ä:

`dct = {"go": 1, "went": 2, "gone": 3}`

–ù–∞–ø–µ—á–∞—Ç–∞–π—Ç–µ —Å–ª–æ–≤–∞—Ä—å

````python
#OK

a = [[n.strip(punctuation).strip('¬´¬ª') for n in i] for i in z]

b = sum(a, [])

c = dict((g, b.count(g)) for g in set(b))
print('–≠—Ç–æ —Å–æ–∑–¥–∞–Ω–Ω—ã–π —Å–ª–æ–≤–∞—Ä—å:', c)

d = sorted(c.values(), reverse = True)
e = {}

for j in d:
    for m in c.keys():
        if c[m] == j:
            e[m] = c[m]

print('–≠—Ç–æ —Å–ª–æ–≤–∞—Ä—å, –≤ –∫–æ—Ç–æ—Ä–æ–º –∫–ª—é—á–∏ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω—ã –ø–æ –∏—Ö –∑–Ω–∞—á–µ–Ω–∏—é –æ—Ç –±–æ–ª—å—à–µ–≥–æ –∫ –º–µ–Ω—å—à–µ–º—É:', e)
````

````python
#OK

print('–í –Ω–∞—à–µ–º —Å–ª–æ–≤–∞—Ä–µ —Å–ª–µ–¥—É—é—â–∏–µ —Å–ª–æ–≤–∞ –ø–æ–≤—Ç–æ—Ä—è—é—Ç—Å—è: ')

for k, v in e.items():
  if v > 1:
    print(f'–°–ª–æ–≤–æ " {k} " –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è {v} —Ä–∞–∑–∞')
````
